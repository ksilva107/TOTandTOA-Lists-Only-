```python
import numpy as np  
import matplotlib.pyplot as plt
def gaus(c,h,w,x):# takes in one set of gaussian parameters:#c= center, h=height, w=width, x=time
    f= h*np.exp(-(x-c)**2/(2*w**2))
    return f,x
def Drawit(lst):# will take in a list of gaussian parameters
    x=[]
    y=[]
    for i in lst[0][1]:
        x.append(i)
    for i in lst[0][0]:
        y.append(i)

    width=0.5*((max(x)-min(x))/(len(x)-1))
    for k in range(len(x)):
        
        for i in range(1,len(lst)):
            for j in range(len(lst[i][0])):
            
                
                if lst[i][1][j]< (x[k]+(width)) and lst[i][1][j]>=(x[k]-(width)):
                    y[k]=y[k]+lst[i][0][j]
                if lst[i][1][j]< min(x)-width:
                    x.append(lst[i][1][j])
                    y.append(lst[i][0][j])
                if lst[i][1][j]> max(x)+width:
                    x.append(lst[i][1][j])
                    y.append(lst[i][0][j])
                    
    plt.plot(x,y)
    plt.xlabel('Time')
    plt.ylabel('Amount of Signal')
    plt.title('Signal vs. Time')
    return np.array(y)
l=gaus(1,1,1,np.linspace(-2,10,60))# gaus(center, height, width, start, stop, number of samples)
m=gaus(5,2,1,np.linspace(-2,10,60))
n=gaus(8,3,1,np.linspace(-2,10,60))
lst2=[l,m,n]
def intersect(t):# input designated y-value for intersection 
    x = np.linspace(-2,10,60)
    f = Drawit(lst2)
    lenf=len(f)
    hlist= np.full((lenf,1),t)
    g = plt.axhline(y=t , xmin=-2, xmax=10, linewidth=1)
    idx = np.argwhere(np.diff(np.sign(f - hlist)))
    
    xlist=[]
    ylist=[]
    xidx=[]
    xlist2=[]
    ylist2=[]
    slopes=[]
    bs=[]
    xs=[]
    ys=[]
    
    for i in range(len(x[idx])):
        if x[idx][i][1] not in xlist:
            xlist.append(x[idx][i][1])
            xidx.append(idx[i][1])
        if f[idx][i][1] not in ylist:
            ylist.append(f[idx][i][1])
    for i in range(len(xidx)):
        xlist2.append(x[xidx[i]+1])
        ylist2.append(f[xidx[i]+1])
    for i in range(len(xlist)):
        slopes.append((ylist2[i]-ylist[i])/(xlist2[i]-xlist[i]))
    for i in range(len(xlist)):
        bs.append(ylist[i]-(slopes[i]*xlist[i]))
    for i in range(len(xlist)):
        xs.append((t-bs[i])/slopes[i])
    for i in range(len(xlist)):
        ys.append(t)
    plt.scatter(xs,ys)
    durations=[]
    starts=[]
    for i in range(len(slopes)):
        if np.sign(slopes[i])== 1 and np.sign(slopes[i+1])==-1:
            durations.append(xs[i+1]-xs[i])
            starts.append(xs[i])
    return (xs,ys,durations, starts)# returns list... 1st= intersection pts, 2nd= y value of intersection (threshold), 3rd= time at or above designated threshold value, 4th=time of arrivals 
intersect(0.7)
```
